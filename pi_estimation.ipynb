```python
# Import the libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
# Function to estimate pi

def estimate_pi(n):
    m = 0
    list_out = []
    list_in = []
    for i in range(n):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if x**2 + y**2 <= 1:
            m += 1
            list_in.append([x,y])
        else:   
            list_out.append([x,y])
    return 4 * m / n, pd.DataFrame(list_in), pd.DataFrame(list_out)
```

```python
# Estimate pi

pi, list_in, list_out = estimate_pi(20000)
print(pi)
```

```python
# Plot the data used for estimation

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5,5))
ax.scatter(list_in[0], list_in[1], s=2)
ax.scatter(list_out[0], list_out[1], s=2)
```
